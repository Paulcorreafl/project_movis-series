```python
import pandas as pd
import numpy as np
```

```python
df_films= pd.read_csv('../csv_clean/csv_clean.csv')
#df_cinema.head()
df_films
```

```python
df_films.info()
```

```python
df_films['series&movie'] = df_films['year'].where(df_films['year'].str.contains('–', na=False))
df_films
```

```python
# Cambiar los valores excepto los nulos (NaN)
df_films['series&movie'] = df_films['series&movie'].apply(lambda x: 'series' if pd.notnull(x) else x)
df_films
```

```python
# Rellenar los valores nulos con el nombre 'series'
df_films['series&movie'] = df_films['series&movie'].fillna('movie')
df_films
```

```python
# Eliminar todas las columnas excepto 'title' y 'series&films'
columnas_a_eliminar = df_films.columns.difference(['title', 'series&movie'])
df_films.drop(columnas_a_eliminar, axis=1, inplace=True)
```

```python
df_films
```

```python
df_films.to_csv('../csv_clean/csv_series&films.csv', index=False, encoding='utf-8')
```
