```python
import pandas as pd
import numpy as np
```

```python
df_films= pd.read_csv('../csv_clean/csv_clean.csv')
#df_cinema.head()
df_films
```

```python
df_films.info()
```

```python
# Extraer solo el primer año
df_films['year'] = df_films['year'].str.extract(r'(\d{4})')

df_films
```

```python
# Eliminar todas las columnas excepto 'title' y 'year'
columnas_a_eliminar = df_films.columns.difference(['title', 'year',])
df_films.drop(columnas_a_eliminar, axis=1, inplace=True)
```

```python
df_films
```

```python
df_films.to_csv('../csv_clean/csv_year.csv', index=False, encoding='utf-8')
```
