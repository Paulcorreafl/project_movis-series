```python
import pandas as pd
import numpy as np
```

```python
df_films= pd.read_csv("../data/n_movies.csv")
#df_cinema.head()
df_films
```

```python
# Eliminar filas con valores nulos
datos_sin_nulos = df_films.dropna()
datos_sin_nulos
```

```python
#eliminar duplicados
df_films = datos_sin_nulos.drop_duplicates(subset=['title'], keep='last')
df_films 
```

```python
df_films.info()
```

```python
df_films.to_csv('../csv_clean/csv_clean.csv', index=False, encoding='utf-8')
```
