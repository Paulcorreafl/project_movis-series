```python
import pandas as pd
import numpy as np
```

```python
df_films= pd.read_csv('../csv_clean/csv_clean.csv')
#df_cinema.head()
df_films
```

```python
df_films.info()
```

```python
# Limpiar y aplanar la columna 'stars'
import ast
df_films['columna'] = df_films['stars'].apply(lambda x: ', '.join([name.strip("', ") for name in ast.literal_eval(x)]))
df_films
```

```python
valores_unicos = df_films['columna'].unique()
valores_unicos
```

```python
# Eliminar todas las columnas excepto 'title' y 'series&films'
columnas_a_eliminar = df_films.columns.difference(['title', 'columna'])
df_films.drop(columnas_a_eliminar, axis=1, inplace=True)
```

```python
df_films
```

```python
df_films.rename(columns={'columna': 'stars'}, inplace=True)
df_films
```

```python
# Eliminar '|Stars:' de la columna 'columna'
df_films['stars'] = df_films['stars'].str.replace(r'\|, Stars:', '', regex=True)
df_films
```

```python
# Eliminar las comas dobles (,,) en la columna 'stars'
df_films['stars'] = df_films['stars'].str.replace(r',\s*,', ',', regex=True)
df_films
```

```python
df_films.to_csv('../csv_clean/csv_stars.csv', index=False, encoding='utf-8')
```
